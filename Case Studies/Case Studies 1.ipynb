```python
hr_data = {}
def insert_data():
    name = str(input("Enter Your Name:"))
    num = int(input("Enter the number of skills you have:"))
    skills = set()
    for i in range(num):
        skill = str(input(f"Emter Skill {i+1}:"))
        skills.add(skill)
    hr_data[name] = skills
```

```python
# Case Study B: Product Inventory Analyzer (Warehouse Comparison)
# Scenario:
# Two warehouses maintain their product details separately, but the management wants to know how their inventories overlap.
# Problem Statement:
# Create a program that accepts product details for two warehouses (Product ID, Product Name, Quantity) from the user and performs the following:
# 1.	Display the common products available in both warehouses (based on product name).
# 2.	Display all unique products across both warehouses.
# 3.	Combine all product details and display the products sorted in descending order of quantity.
```

```python
# Warehouse 1 Inventory
warehouse1 = [
    {"id": 101, "name": "Laptop", "quantity": 50},
    {"id": 102, "name": "Mouse", "quantity": 200},
    {"id": 103, "name": "Keyboard", "quantity": 150},
    {"id": 104, "name": "Monitor", "quantity": 80},
    {"id": 105, "name": "Printer", "quantity": 40}
]

# Warehouse 2 Inventory
warehouse2 = [
    {"id": 201, "name": "Laptop", "quantity": 60},
    {"id": 202, "name": "Mouse", "quantity": 180},
    {"id": 203, "name": "Webcam", "quantity": 90},
    {"id": 204, "name": "Monitor", "quantity": 70},
    {"id": 205, "name": "Scanner", "quantity": 30}
]
```

```python
def insert_data(warehouse):
    num = int(input("Enter the no of products:"))
    for i in range(num):
        id = int(input("Enter the Product id"))
        name = str(input("Enter the Product Name"))
        quantity = int(input("Enter the Quantity"))
        data = {"id":id, "name":name, "quantity":quantity}
        warehouse.append(data)
```

```python
# displaying common products in both warehouses based on name
products1 = set()
products2 = set()

for product in warehouse1:
    products1.add(product["name"])
for product in warehouse2:
    products2.add(product["name"])

print("Common Products:",products1 & products2)
```

```python
# all unique products
print("Unique Products:",products1 | products2)
```

```python
# sorting
combined = warehouse1 + warehouse2
combined.sort(key=lambda x: x["quantity"], reverse=True)

print("\nProducts sorted by quantity:")
for p in combined:
    print(p)
```

```python
# Case Study C: University Course Registration System
# Scenario:
# The university tracks student course registrations, but wants a summary of course participation.
# Problem Statement:
# Write a program where the user enters student names and the courses they have registered for (store them as tuples). Then perform:
# 1.	Identify all students enrolled in both ‘Data Science’ and ‘AI’.
# 2.	Find the total number of unique courses offered.
# 3.	Find all courses that no student has repeated.
```

```python
students = [
    ("Amit", {"AI", "Data Science", "ML"}),
    ("Neha", {"AI", "Cloud"}),
    ("Ravi", {"Data Science", "DBMS"})
]

print("Students enrolled in both AI and Data Science:")
for name, courses in students:
    if "AI" in courses and "Data Science" in courses:
        print(name)

all_courses = set()
for _, courses in students:
    all_courses.update(courses)

print("\nTotal number of unique courses:")
print(len(all_courses))

course_count = {}
for _, courses in students:
    for c in courses:
        course_count[c] = course_count.get(c, 0) + 1

print("\nCourses not repeated by any student:")
for c, count in course_count.items():
    if count == 1:
        print(c)
```

```python
# Case Study D: Customer Purchase Insights (E-commerce Scenario)
# Scenario:
# An e-commerce website tracks customer purchases daily. The company wants to derive insights about purchasing patterns.
# Problem Statement:
# Create a Python program where the user inputs daily customer purchase details as tuples — each containing (Customer Name, List of Products Purchased). Perform:
# 1.	Display the total number of unique products bought by all customers combined.
# 2.	Find customers who purchased both ‘Laptop’ and ‘Headphones’.
# 3.	Identify and display the most frequently purchased product(s) among all customers.
```

```python
customers = [
    ("Rahul", ["Laptop", "Mouse", "Headphones"]),
    ("Sneha", ["Mobile", "Headphones"]),
    ("Aman", ["Laptop", "Headphones"])
]

all_products = set()
for _, items in customers:
    all_products.update(items)

print("Total unique products purchased:")
print(all_products)

print("\nCustomers who purchased Laptop and Headphones:")
for name, items in customers:
    if "Laptop" in items and "Headphones" in items:
        print(name)

product_frequency = {}
for _, items in customers:
    for p in items:
        product_frequency[p] = product_frequency.get(p, 0) + 1

max_freq = max(product_frequency.values())

print("\nMost frequently purchased product(s):")
for p, freq in product_frequency.items():
    if freq == max_freq:
        print(p)
```

```python
# Case Study E: Travel Planner (Data Deduplication Challenge)
# Scenario:
# A travel company collected booking data from multiple agents, but the data contains duplicates and unorganized entries.
# Problem Statement:
# Develop a Python program that accepts multiple travel booking details (Traveler_Name, Destination, Date) from the user and performs the following operations:
# 1.	Remove duplicate records using sets.
# 2.	Display all unique destinations from the bookings.
# 3.	Find and display the top 3 most visited destinations based on their frequency.
```

```python
bookings = [
    ("Amit", "Goa", "2024-05-10"),
    ("Neha", "Manali", "2024-06-15"),
    ("Amit", "Goa", "2024-05-10"),
    ("Ravi", "Goa", "2024-07-01")
]

unique_bookings = set(bookings)
print("Unique booking records:")
for b in unique_bookings:
    print(b)

destinations = []
for _, dest, _ in unique_bookings:
    destinations.append(dest)

print("\nUnique destinations:")
print(set(destinations))

from collections import Counter
top_destinations = Counter(destinations).most_common(3)

print("\nTop 3 most visited destinations:")
for d, count in top_destinations:
    print(d, "visited", count, "times")
```

```python
# Case Study F: Student Attendance Analyzer
# Scenario:
# A class teacher keeps daily attendance records, but wants to analyze attendance trends.
# Problem Statement:
# Write a program where the user enters student names for five different days (store each day’s attendance as a set). Then perform:
# 1.	Find students who were present every day.
# 2.	Find students who were absent on at least one day.
# 3.	Display the total unique students across all five days.
```

```python
day1 = {"Amit", "Neha", "Ravi"}
day2 = {"Amit", "Ravi"}
day3 = {"Amit", "Neha"}
day4 = {"Amit", "Neha", "Ravi"}
day5 = {"Amit", "Ravi"}

days = [day1, day2, day3, day4, day5]

present_all_days = set.intersection(*days)
print("Students present every day:")
print(present_all_days)

all_students = set.union(*days)
absent_some_day = all_students - present_all_days

print("\nStudents absent on at least one day:")
print(absent_some_day)

print("\nTotal unique students:")
print(len(all_students))
```

```python
# Case Study G: Movie Recommendation Tracker
# Scenario:
# An OTT platform wants to analyze user movie preferences to recommend similar content.
# Problem Statement:
# Write a Python program where users input their names and a tuple of favorite movies. Then:
# 1.	Find users who share at least two movies in common with another user.
# 2.	Display all unique movies in the database.
# 3.	Identify movies that only one user has marked as favorite.
```

```python
users = {
    "Amit": {"Inception", "Interstellar", "Matrix"},
    "Neha": {"Matrix", "Inception", "Avatar"},
    "Ravi": {"Avatar", "Titanic"}
}

print("Users sharing at least two common movies:")
for u1 in users:
    for u2 in users:
        if u1 != u2:
            if len(users[u1].intersection(users[u2])) >= 2:
                print(u1, "and", u2)

all_movies = set()
for movies in users.values():
    all_movies.update(movies)

print("\nAll unique movies:")
print(all_movies)

movie_count = {}
for movies in users.values():
    for m in movies:
        movie_count[m] = movie_count.get(m, 0) + 1

print("\nMovies liked by only one user:")
for m, count in movie_count.items():
    if count == 1:
        print(m)
```

```python
# Case Study H: Sports Event Participation Summary
# Scenario:
# A sports club records data for participants in multiple events. Each participant can register for multiple sports.
# Problem Statement:
# Write a Python program to take inputs as (Participant_Name, [List_of_Sports]) and perform:
# 1.	Display participants who are taking part in both ‘Cricket’ and ‘Football’.
# 2.	Find the total number of unique sports being played.
# 3.	Create a tuple containing sports with the highest participation count.
```

```python
participants = {
    "Amit": {"Cricket", "Football"},
    "Neha": {"Badminton", "Cricket"},
    "Ravi": {"Cricket", "Football"},
    "Sneha": {"Tennis"}
}

print("Participants playing both Cricket and Football:")
for name, sports in participants.items():
    if "Cricket" in sports and "Football" in sports:
        print(name)

all_sports = set()
for sports in participants.values():
    all_sports.update(sports)

print("\nTotal unique sports:")
print(len(all_sports))

sport_count = {}
for sports in participants.values():
    for s in sports:
        sport_count[s] = sport_count.get(s, 0) + 1

max_count = max(sport_count.values())

print("\nSport(s) with highest participation:")
for s, count in sport_count.items():
    if count == max_count:
        print(s)
```
