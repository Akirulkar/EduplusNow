# Weather Data Analysis System

## Given Weather Data

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature = [32, 35, 38, 36, 34, 39, 33]
rainfall = [0, 5, 12, 0, 20, 45, 0]
humidity = [45, 50, 55, 60, 65, 70, 48]
```

### 1. Heatwave Detection (Temperature > 37°C)

```python
for i in range(len(days)):
    if temperature[i] > 37:
        print(days[i])
```

### 2. Heavy Rainfall Days (Rainfall ≥ 20 mm)

```python
for i in range(len(days)):
    if rainfall[i] >= 20:
        print(days[i])
```

### 3. High Humidity Alert (Humidity > 65%)

```python
for i in range(len(days)):
    if humidity[i] > 65:
        print(days[i])
```

### 4. Dry Days (Rainfall == 0 mm)

```python
for i in range(len(days)):
    if rainfall[i] == 0:
        print(days[i])
```

### 5. Comfortable Weather (Temp ≤ 35°C and Humidity < 60%)

```python
for i in range(len(days)):
    if temperature[i] <= 35 and humidity[i] < 60:
        print(days[i])
```

### 6. Extreme Weather (Temp > 37°C or Rainfall > 30 mm)

```python
for i in range(len(days)):
    if temperature[i] > 37 or rainfall[i] > 30:
        print(days[i])
```

### 7. Rainy but Cool Days (Rainfall > 10 mm and Temp < 35°C)

```python
for i in range(len(days)):
    if rainfall[i] > 10 and temperature[i] < 35:
        print(days[i])
```

### 8. Most Humid Day

```python
max_humidity = max(humidity)
index = humidity.index(max_humidity)
print(days[index], max_humidity)
```

### 9. Temperature Comparison (Monday vs Saturday)

```python
if temperature[0] > temperature[5]:
    print("Monday is hotter")
else:
    print("Saturday is hotter")
```

### 10. Climate Trend Warning (Temp != Average Temp)

```python
avg_temp = sum(temperature) / len(temperature)
for i in range(len(days)):
    if temperature[i] != avg_temp:
        print(days[i])
```
