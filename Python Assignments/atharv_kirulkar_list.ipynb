```python
#1. Create a list of 10 numbers and print the list.
ten_numbers = [x for x in range(1,11)]
print(ten_numbers)
```

```python
#2. Create a list containing different data types (integer, float, string, boolean) and print its type.
diff_data_types = [1,"Hello",2.4,True]
print(type(diff_data_types))
```

```python
#3. Access and print the first, middle, and last elements from a list of your choice.
n = len(ten_numbers)

if n%2 == 0:
    middle_number = []
    middle_number.append(ten_numbers[n//2-1])
    middle_number.append(ten_numbers[n//2])
else:
    middle_number = ten_numbers[n//2]
    
print(ten_numbers[0])
print(middle_number)
print(ten_numbers[n-1])
```

```python
#4. Slice the list to print the first 3 elements and last 3 elements separately.
print(ten_numbers[:3])
print(ten_numbers[:-4:-1])
```

```python
#5. Replace the second element of a list with a new value.
diff_data_types[1]="hey"
print(diff_data_types)
```

```python
#6. Add a new element at the end of a list using append() and display the updated list.
diff_data_types.append(False)
print(diff_data_types)
```

```python
#7. Insert an element at index 2 using insert() and print the updated list.
diff_data_types.insert(2,"why")
print(diff_data_types)
```

```python
#8. Extend the list with another list of your choice and display the final list.
fruits = ["orange","banana","kiwi","apple"]
diff_data_types.extend(fruits)
print(diff_data_types)
```

```python
#9. Remove a specific element by value using remove() and display the result.
diff_data_types.remove('apple')
print(diff_data_types)
```

```python
#10. Remove the element at index 3 using pop() and display which element was removed.
removed=diff_data_types.pop(3)
print(diff_data_types)
print(removed)
```

```python
#11. Delete the element at index 0 using del keyword and print the resulting list.
del diff_data_types[0]
print(diff_data_types)
```

```python
#12. Clear all elements from the list using clear() and print the empty list.
diff_data_types.clear()
print(diff_data_types)
```

```python
#13. Recreate a list and check whether a specific element (for example "apple") exists in the list using in keyword.
print("apple" in fruits)
```

```python
#14. Find the index position of a specific element in the list using the index() method.
print(fruits.index("apple"))
```

```python
#15. Count how many times a specific value appears in the list using count().
print(fruits.count("apple"))
```

```python
#16. Sort a list of numbers in ascending order using sort() and display the result.
fruits.sort()
print(fruits)
```

```python
#17. Sort a list of strings in descending order using sort(reverse=True) and display the result.
fruits.sort(reverse=True)
print(fruits)
```

```python
#18. Reverse the order of a list using reverse() and print the updated list.
fruits.reverse()
print(fruits)
```

```python
#19. Copy the contents of one list to another using the copy() method and modify one list to see the effect.
cars = ["suzuki","tata","mahindra","skoda"]
fruits=cars.copy()
print(fruits)
```

```python
#20. Combine the two lists using the + operator and print the combined list.
fruits = ["orange","banana","kiwi","apple"]
combined = fruits + cars
print(combined)
```
