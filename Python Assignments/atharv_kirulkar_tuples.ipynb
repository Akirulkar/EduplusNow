```python
#8. Use sum() to calculate the total sales from a tuple of daily sales data.
daily_sales_data=(20,30,50,100,60,10,15,35)
print("total sales:",sum(daily_sales_data))
```

```python
#9. Sort a tuple of numbers in ascending order using sorted() and convert it back to a tuple,
numbers_sorted = tuple(sorted(numbers))
print(numbers_sorted)
```

```python
#10. Sort a tuple of numbers in descending order using sorted() with reverse-True.
numbers_sorted = tuple(sorted(numbers, reverse=True))
print(numbers_sorted)
```

```python
#15. Create a tuple from a range of numbers 1 to 10 using the tuple() and range() functions.
numbers = tuple(range(1,11))
print(numbers)
```

```python
#16. Use any() on a tuple of hoolean values to check if at least one element is True.
binary = (1,0,1,0)
print(any(binary))
```

```python
#18. Concatenate two tuples representing first names and last names.
first_name = ("Shrikant","Arvind","JK")
last_name = ("Tiwari","Choudhary","Talpade")
names = tuple(first_name + last_name)
print(names)
```

```python
#19. Repeat a tuple containing ("Al", "ML") three times using the operator.
tuple1 = ("Al", "ML")
tuple1x3 = tuple(tuple1*3)
print(tuple1x3)
```

```python
#20. Combine two tuples (names and scores) using zip() and convert the result into a tuple.
names = ("Shrikant","Arvind","JK")
scores = (20,15,25)
combined = tuple(zip(names,scores))
print(combined)
```

```python
#21. Reverse a tuple using the reversed() function and convert it back to a tuple.
scores_rev = tuple(reversed(scores))
print(scores_rev)
```

```python
#24. Create a tuple and demonstrate immutability by trying to modify an element (observe the error).
names = ("Shrikant","Arvind","JK")
names[1] = "Moosa"
```

```python
#25. Compare the memory size of a list and a tuple containing the same elements using sys.getsizeof().
import sys

names_tuple = ("Shrikant","Arvind","JK")
names_list = ["Shrikant","Arvind","JK"]
print(sys.getsizeof(names_tuple))
print(sys.getsizeof(names_list))
```

```python
#26. Create a tuple of 10 numbers and calculate the average using sum() and len().
numbers = tuple(range(1,11))
print(numbers)
print("Average:",sum(numbers)/len(numbers))
```

```python
#27. Create a tuple of city names and print them in reverse order using slicing.
cities = ("Pune","Kolhapur","Satara","Nagpur","Karad","Ratnagiri")
print("Reverse order:",cities[::-1])
```

```python
#28. Convert a tuple of numbers into a list, append a new number, and convert it back into a tuple.
numbers_list = list(numbers)
print(numbers_list)
numbers_list.append(11)
numbers = tuple(numbers_list)
print(numbers)
```

```python
#29. Create a tuple of employee details (Name, Age, Department) and use unpacking to print each value separately.
employees = ("Shrikanr","45","Agent")
Name, Age, Department = employees
print("Name:", Name)
print("Age:",Age)
print("Department:",Department)
```

```python
#30. Create two tuples (product name, price) and combine them using zip(); then display them sorted by price in descending order
product_name = ("Chakli","Chiwda","Ladoo")
prices = (600,500,800)
combined = tuple(zip(prices,product_name))
print(combined)
print(tuple(sorted(combined, reverse = True)))
```
