```python
import pandas as pd
```

```python
# 1. Create a Pandas Series using a list of 5 random numbers and custom index names.Then print the series and its data type.
df = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
print(df)
```

```python
# 2. Create a DataFrame with columns: "Student_Name", "Maths", "Science", "English". Fill it with 5 sample student records and display the top 3 rows using head().
data = {
    'Student_Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Maths': [85, 90, 78, 92, 88],
    'Science': [79, 88, 85, 95, 80],
    'English': [92, 75, 80, 88, 90]
}

df_students = pd.DataFrame(data)
print(df_students.head(3))
```

```python
# 3. Import a CSV file (use any sample dataset) into a DataFrame. Display the number of rows and columns using shape, and print column names.
csv_path = '/content/sample_data/mnist_train_small.csv'
df_csv = pd.read_csv(csv_path)
print("Shape: ", df_csv.shape)
print("Columns: ", df_csv.columns)
```

```python
# 4. Add a new column named "Total_Marks" that sums all three subject scores. Then create another column "Average" by dividing Total_Marks by 3.
df_students['Total_Marks'] = df_students['Maths'] + df_students['Science'] + df_students['English']
df_students['Average'] = df_students['Total_Marks'] / 3
print(df_students)
```

```python
# 5. Introduce a missing value (NaN) in the "Science" column and then:
#     (a) Identify missing values using isnull()
#     (b) Replace the missing value with the column mean using fillna().
df_students.loc[3,'Science'] = None
print(df_students)
print(df_students.isnull())
df_students['Science'].fillna(df_students['Science'].mean(), inplace=True)
print(df_students)
```

```python
# 6. Filter and display all students whose "Average" score is greater than 75.
df_students_filtered = df_students[df_students['Average'] > 75]
print(df_students_filtered)
```

```python
# 7. Sort the DataFrame by "Total_Marks" in descending order and show the top 3 performers.
df_students_sorted = df_students.sort_values(by='Total_Marks', ascending=False)
print(df_students_sorted.head(3))
```

```python
# 8. Group students based on their "English" score range (like above or below 70) and find the mean of "Maths" and "Science" marks for each group.
df_students['English_Range'] = pd.cut(df_students['English'], bins=[0, 70, 100], labels=['Below 70', 'Above 70'])
print(df_students)
print(df_students.groupby('English_Range')[['Maths', 'Science']].mean())
```

```python
# 9. Merge two DataFrames:
#     (a) One with "Student_Name" and "Total_Marks"
#     (b) Another with "Student_Name" and "Sports_Score"
#     Then create a new column "Overall_Score" = Total_Marks + Sports_Score.
df1 = pd.DataFrame({'Student_Name': df_students['Student_Name'], 'Total_Marks': df_students['Total_Marks']})
df2 = pd.DataFrame({'Student_Name': df_students['Student_Name'], 'Sports_Score': [100, 95, 88, 70, 60]})
df_merged = pd.merge(df1, df2, on='Student_Name')
df_merged['Overall_Score'] = df_merged['Total_Marks'] + df_merged['Sports_Score']
print(df_merged)
```

```python
# 10. Export the final DataFrame to a CSV file on your Desktop (use os library to define path) and verify if the file is successfully created using os.path.exists().
csv_path = '/content/sample_data/final.csv'
df_merged.to_csv(csv_path, index=False)
import os
os.path.exists(csv_path)
```
