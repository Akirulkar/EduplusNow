```python
# 2. Download all required NLTK resources (punkt, stopwords, wordnet, averaged_perceptron_tagger_eng) using nltk.download(). 
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# 3. Write a Python program to tokenize a given paragraph into sentences using sent_tokenize(). 
from nltk.tokenize import word_tokenize, sent_tokenize
paragraph = '''Piranhas rarely feed on large animals; they eat smaller fish and aquatic plants. When confronted with humans, piranhas’ first instinct is to flee, not attack. Their fear of humans makes sense. Far more piranhas are eaten by people than people are eaten by piranhas. If the fish are well-fed, they won’t bite humans.'''
print(paragraph)
sentence = sent_tokenize(paragraph)
print(sentence)
```

```python
# 4. Tokenize the same paragraph into individual words using word_tokenize(). 
words = word_tokenize(paragraph)
print(words)
```

```python
# 5. Remove stopwords from a given text using NLTK stopwords list. 
from nltk.corpus import stopwords

stopwords = set(stopwords.words('english'))
filtered_words = [x for x in words if x.lower() not in stopwords]
print(filtered_words)
```

```python
# 6. Perform stemming on a list of words using the PorterStemmer and print the output. 
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(w) for w in filtered_words]
print(stemmed_words)
```

```python
# 7. Perform lemmatization on the same list of words using WordNetLemmatizer and compare results with stemming. 
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmarized_words = [lemmatizer.lemmatize(w) for w in filtered_words]
print(lemmarized_words)
```

```python
# 8. Apply Part of Speech (POS) tagging to a given sentence and print each word with its tag. 
pos_tags = nltk.pos_tag(filtered_words)
for tag in pos_tags:
    print(tag)
```

```python
# 9. Perform frequency distribution on a paragraph and print the top 10 most frequent words. 
from nltk import FreqDist
freq_dist = FreqDist(filtered_words)
print("Top 10 most frequent words")
for word, freq in freq_dist.most_common(10):
    print(f"{word} : {freq}")
```

```python
# 10. Plot a graph of the top 15 most frequent words using FreqDist().plot(). 
freq_dist.plot(15, title = 'Top 15 most frequent words')
```

```python
# 11. Perform Named Entity Recognition (NER) on a text containing names of people, places, and organizations. 

from nltk import ne_chunk

nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

text = """
Elon Musk visited the Tesla headquarters in Palo Alto last Tuesday. 
He then flew to Berlin to meet with officials from the European Union 
to discuss future expansion plans for SpaceX.
"""
tokens = word_tokenize(text)
tags = nltk.pos_tag(tokens)

tree = ne_chunk(tags, binary=False)
for chunk in tree:
    if hasattr(chunk, 'label'):
        entity_name = " ".join(c[0] for c in chunk)
        entity_label = chunk.label()
        print(f"{entity_name} : {entity_label}")
```

```python
# 12. Create a function that accepts text input and returns a list of nouns and verbs separately using POS tagging. 
def show_n_v(text):
    words = word_tokenize(text)
    pos_tags = nltk.pos_tag(words)
    nouns = []
    verbs = []
    for tag in pos_tags:
        if tag[1] == 'NN':
            nouns.append(tag)
            continue
        if tag[1] == 'VB':
            verbs.append(tag)
    return nouns,verbs
nouns, verbs = show_n_v(text)
print('Nouns: ', nouns)
print('Verbs: ', verbs)
```

```python
# 13. Count the number of adjectives and adverbs present in a given text. 
def count_adj_adv(text):
    words = word_tokenize(text)
    pos_tags = nltk.pos_tag(words)
    adj = 0
    adv = 0
    for tag in pos_tags:
        if tag[1] == 'JJ':
            adj += 1
            continue
        if tag[1] == 'RB':
            adv += 1
    return adj, adv
adj, adv = count_adj_adv(text)
print(f"No of Adjectives: {adj}")
print(f"No of Adverbs: {adv}")
```

```python
# 16. Compare and print the difference between stemming and lemmatization for a list of 10 verbs. 
verbs = ["build", "analyze", "process", "deliver", "code", "optimize", "update", "manage", "extract", "debug"]
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(x) for x in verbs]
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(x) for x in verbs]
print("Verb\tStemmed\tLemmatized")
for i in range(len(verbs)):
    print(f"{verbs[i]}\t{stemmed_words[i]}\t{lemmatized_words[i]}")
```

```python
# 17. Extract all bigrams (two-word combinations) from a sentence using nltk.bigrams(). 
sentence_bi = "The quick brown fox jumps over the lazy dog"
tokens = word_tokenize(sentence_bi)
list_bigrams = list(nltk.bigrams(tokens))
print(list_bigrams)
```

```python
# 18. Create a frequency distribution of bigrams in a given paragraph.
paragraph_bi = """Data science is an interdisciplinary field. Data science uses 
scientific methods, processes, algorithms and systems to extract knowledge."""
tokens = word_tokenize(paragraph_bi)
bi = nltk.bigrams(tokens)
f_dist_bi = FreqDist(bi)
for bi, f in f_dist_bi.most_common(3):
    print(f"Bigram: {bi}, Frequency: {f}")
```

```python
# 19. Write a Python function that removes punctuation marks before tokenization using regex and NLTK.
import re
def clean_and_tokenize(text):
    clean_text = re.sub(r'[^\w\s]','',text)
    tokens = word_tokenize(clean_text)
    return tokens
print(f"Tokens using simple tokenize: \n {tokens}")
tokens = clean_and_tokenize(paragraph_bi)
print(f"Tokens using clean_and_tokenize: \n {tokens}")
```

```python
# 20. Perform sentiment analysis on a small list of sentences using the NLTK movie_reviews dataset. 

import nltk
from nltk.corpus import movie_reviews
import random

pos_ids = movie_reviews.fileids('pos')
neg_ids = movie_reviews.fileids('neg')

documents = [(list(movie_reviews.words(fid)), 'pos') for fid in pos_ids[:500]]
documents += [(list(movie_reviews.words(fid)), 'neg') for fid in neg_ids[:500]]

random.shuffle(documents)

all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words() if w.isalpha())
word_features = list(all_words)[:2000]

def find_features(document):
    words = set(document)
    return {word: (word in words) for word in word_features}

featuresets = [(find_features(rev), category) for (rev, category) in documents]
# 19. Write a Python function that removes punctuation marks before tokenization using regex and NLTK. 
classifier = nltk.NaiveBayesClassifier.train(featuresets)

sample = [
    "This movie was an absolute masterpiece of cinema.",
    "I hated every minute of this boring film.",
]

for sentence in sample:
    tokens = sentence.lower().split()
    feats = find_features(tokens)
    prob_dist = classifier.prob_classify(feats)
    print(f"Sentence: {sentence}")
    print(f"Pos: {prob_dist.prob('pos'):.4f} | Neg: {prob_dist.prob('neg'):.4f}")
    print(f"Result: {classifier.classify(feats)}\n")
```

```python
# 24. Prepare feature sets using the movie_reviews dataset and split them into training and testing data. 
featuresets = [(find_features(rev), category) for (rev, category) in documents]
train_set, test_set = featuresets[50:], featuresets[:50]
```
