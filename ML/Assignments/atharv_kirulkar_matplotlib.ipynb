```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Q1: Create a simple line plot of y = 2x + 3 for x values ranging from 0 to 10.
x = np.linspace(0,10,100)
y = (2*x) + 3

plt.plot(x, y, 'r-')

plt.title("Simple Line Plot")
plt.xlabel("x")
plt.ylabel("y = 2x + 3")

plt.show()
```

```python
# Q2: Plot sin(x) and cos(x) on the same graph with different colors, line styles, and labels, and display a legend.
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'ro--', label = 'sin(x)')
plt.plot(x, y2, 'bx-', label = 'cos(x)')

plt.title("sin(x) & cos(x) vs x")
plt.xlabel("x values")
plt.ylabel("y values")

plt.legend()
plt.show()
```

```python
# Q3: Create a bar chart showing the sales of four products: A, B, C, and D with random values.
products = ['A', 'B', 'C', 'D']
sales = [10000, 2000, 30000, 15000]

plt.bar(products, sales, color = 'purple')

plt.title('Bar Chart')
plt.xlabel('Products')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```

```python
# Q4: Generate a scatter plot with random x and y values between 0 and 1, and use different colors for points.
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha = 0.3, cmap='viridis')

plt.title("Scatter plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.colorbar()
plt.show()
```

```python
# Q5: Plot a histogram for 1000 random numbers generated from a normal distribution.
x = np.random.randn(1000)

plt.hist(x, bins = 40, color = 'green', edgecolor = 'black', alpha = 0.5)
plt.title("Histogram")
plt.xlabel("Data Range")
plt.ylabel("Freq")

plt.show()
```

```python
# Q6: Draw a pie chart representing the percentage of five fruits: Apple, Banana, Mango, Grapes, and Orange.
fruits = ['Apple', 'Banana', 'Mango', 'Grapes','Orange']
values = [20,23,14,40,30]
colors = ['red', 'yellow','orange','green', 'brown']

plt.pie(values, labels = fruits, colors = colors, autopct = "%1.f%%")
plt.title("Pie Chart")
```

```python
# Q7: Create two subplots in one figure â€“ one showing a sine wave and the other showing a cosine wave.
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x, y1, 'r-')
plt.title("Sine Wave")
plt.xlabel("x values")
plt.ylabel("sin(x)")

plt.subplot(1,2,2)
plt.plot(x, y2, 'b-')
plt.title("Cosine Wave")
plt.xlabel("x values")
plt.ylabel("cos(x)")

plt.suptitle("Two subplots: Sine and Cosine Waves")
plt.tight_layout()
plt.show()
```

```python
# Q8: Customize a line plot by adding title, axis labels, grid lines, and saving it as 'line_plot.png'.
x = np.linspace(0,10,50)
y = x**2

plt.plot(x, y, color = 'red', linestyle = '--', marker = 'o', label = 'x^2')

plt.title("Customised Line Plot")
plt.xlabel("X")
plt.ylabel("X^2")

plt.grid(True)
plt.legend()

plt.savefig("tan(x).png", dpi = 300, bbox_inches='tight')
plt.show()
```

```python
# Q9: Create a 2x2 grid of subplots showing different plots: line, bar, scatter, and histogram.
fig, axis = plt.subplots(2,2, figsize=(10,8))

#axis[0,0]
x = np.linspace(0,10,100)
y = (2*x) + 3

axis[0,0].plot(x, y, 'r-')

axis[0,0].set_title("Simple Line Plot")
axis[0,0].set_xlabel("x")
axis[0,0].set_ylabel("y = 2x + 3")


#axis[0,1]
products = ['A', 'B', 'C', 'D']
sales = [10000, 2000, 30000, 15000]

axis[0,1].bar(products, sales, color = 'purple')

axis[0,1].set_title('Bar Chart')
axis[0,1].set_xlabel('Products')
axis[0,1].set_ylabel('Sales')


#axis[1,0]
x1 = np.random.rand(100)
y1 = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

axis[1,0].scatter(x1, y1, c=colors, s=sizes, alpha = 0.3, cmap='viridis')

axis[1,0].set_title("Scatter plot")
axis[1,0].set_xlabel("X values")
axis[1,0].set_ylabel("Y values")


#axis[1,1]
x2 = np.random.randn(1000)

axis[1,1].hist(x2, bins = 40, color = 'green', edgecolor = 'black', alpha = 0.5)
axis[1,1].set_title("Histogram")
axis[1,1].set_xlabel("Data Range")
axis[1,1].set_ylabel("Freq")

plt.suptitle("Mini DashBoard", fontsize=14)
plt.tight_layout()

plt.show()
```

```python
# Q10: Create a line plot for y = e^x for x between 0 and 5, and change the color, marker, and line style.
x = np.linspace(0,5,100)
y = np.exp(x)

plt.plot(x, y, 'ro--')
plt.title("Exponential line plot")
plt.xlabel("x values")
plt.ylabel("y = e^x")

plt.show()
```
